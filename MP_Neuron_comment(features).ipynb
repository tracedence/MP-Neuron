```python
import sklearn.datasets
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
# getting dataset from sklearn
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
data = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
data['class'] = breast_cancer.target
```

```python
# input
# dropping target column
X = data.drop('class', axis=1)
# traget column
Y = data['class']
```

```python
print(X.shape)
print(Y.shape)
```

```python
X.head()
```

```python
# split the data into training and test dataset
# test_size parameter define the ratio of train and test data
# stratify split the traget variable in approx equal proportion in train and test data
# random_state is used to set the seed of randomly generated numbers. With parameter we will always get same train and test split.
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, stratify = Y, random_state = 23)
```
